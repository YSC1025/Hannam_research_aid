### **뇌졸중 가중치 및 결과 구하기**





```python
# 엑셀 파일 불러오기
file_path = '/content/뇌졸중_가중치.xlsx'

data = pd.read_excel(file_path)
data
```

```python
# 각 열의 가중치를 정의합니다.
weights = {
    '뇌졸중(중풍) 조기증상 인지_팔다리 힘빠짐(35%)': 35,
    '뇌졸중(중풍) 조기증상 인지_말이 어눌해짐(30%)': 30,
    '뇌졸중(중풍) 조기증상 인지_한쪽 눈 안보임(15%)': 15,
    '뇌졸중(중풍) 인지_중심잡기 힘듦(10%)': 10,
    '뇌졸중(중풍) 조기증상 인지_심한 두통(10%)': 10
}

# 가중치를 곱하고 합산하여 '가중치' 열에 값을 추가합니다.
data['가중치'] = data.apply(lambda row: sum(row[col] * weights[col] for col in weights), axis=1)

# 새로운 기준에 따라 결과를 분류합니다.
def classify_result(weight):
    if weight <= 25:
        return "양호"
    elif weight <= 50:
        return "주의"
    elif weight <= 75:
        return "우기"
    else:
        return "심각"

data['결과'] = data['가중치'].apply(classify_result)

# 결과 확인
print(data)
```

```python
data['결과'].unique()
```

```python
# 각 결과의 비율을 계산합니다.
total_count = len(data)  # 전체 데이터 행 수
result_counts = data['결과'].value_counts()  # 각 결과의 발생 횟수
result_percentages = (result_counts / total_count) * 100  # 비율 계산

# 결과 출력
print(result_percentages)
```

### **결과에 따른 남녀 인원수 체크 (뇌졸중)**

```python
# 엑셀 파일 불러오기
file_path = '/content/뇌졸중_결과분류.xlsx'

data = pd.read_excel(file_path)
data
```

```python
import pandas as pd

# 데이터에 가중치와 결과 추가 (이미 수행된 단계)
weights = {
    '뇌졸중(중풍) 조기증상 인지_팔다리 힘빠짐(35%)': 35,
    '뇌졸중(중풍) 조기증상 인지_말이 어눌해짐(30%)': 30,
    '뇌졸중(중풍) 조기증상 인지_한쪽 눈 안보임(15%)': 15,
    '뇌졸중(중풍) 인지_중심잡기 힘듦(10%)': 10,
    '뇌졸중(중풍) 조기증상 인지_심한 두통(10%)': 10
}

data['가중치'] = data.apply(lambda row: sum(row[col] * weights[col] for col in weights), axis=1)

def classify_result(weight):
    if weight <= 25:
        return "양호"
    elif weight <= 50:
        return "주의"
    elif weight <= 75:
        return "우기"
    else:
        return "심각"

data['결과'] = data['가중치'].apply(classify_result)

# 성별에 따른 결과 분류
result_by_gender = data.groupby(['결과', '성별']).size().unstack(fill_value=0)

# 남자와 여자의 결과 출력 (남자: 1, 여자: 0)
for result in result_by_gender.index:
    male_count = result_by_gender.loc[result, 1]
    female_count = result_by_gender.loc[result, 0]
    print(f'{result} 결과:')
    print(f'  남자: {male_count}명')
    print(f'  여자: {female_count}명')
```

### **검진군/비검진군에 따른 남녀 인원수 체크 (뇌졸중)**

```python
# 엑셀 파일 불러오기
file_path = '/content/뇌졸중_결과분류.xlsx'

data = pd.read_excel(file_path)
data
```

```python
import pandas as pd

# '검진군/비검진군 판단'을 기준으로 데이터 나누기 (0: 검진군, 1: 비검진군)
검진군 = data[data['검진군/비검진군 판단'] == 0]
비검진군 = data[data['검진군/비검진군 판단'] == 1]

# 결과별로 성별 인원수를 계산하는 함수
def calculate_gender_counts(group):
    return group.groupby(['결과', '성별']).size().unstack(fill_value=0)

# 검진군과 비검진군에서의 결과별 성별 인원수 계산
검진군_counts = calculate_gender_counts(검진군)
비검진군_counts = calculate_gender_counts(비검진군)

# 결과별로 출력
print('검진군:')
for result in 검진군_counts.index:
    print(f'{result} 결과:')
    print(f'  남자: {검진군_counts.loc[result, 1]}명')
    print(f'  여자: {검진군_counts.loc[result, 0]}명')

print('\n비검진군:')
for result in 비검진군_counts.index:
    print(f'{result} 결과:')
    print(f'  남자: {비검진군_counts.loc[result, 1]}명')
    print(f'  여자: {비검진군_counts.loc[result, 0]}명')
```
