### **컬럼별 빈도수**

```python
import pandas as pd

# 데이터 변환: 수치형 데이터로 변환할 열들
columns_to_convert = ['가구소득_(월)금액', '걷기 실천 일수']
for col in columns_to_convert:
    data[col] = pd.to_numeric(data[col], errors='coerce')

# 모든 변수의 빈도수 계산 및 출력
for col in data.columns:
    print(f'{col}의 값별 빈도수:')

    # 수치형 데이터의 경우
    if pd.api.types.is_numeric_dtype(data[col]):
        value_counts = data[col].value_counts(dropna=False)

    # 범주형 데이터의 경우
    else:
        value_counts = data[col].value_counts(dropna=False)

    print(value_counts, '\n')
```

### **컬럼별 남자와 여자 빈도수**

```python
import pandas as pd

# 데이터 변환: 수치형 데이터로 변환할 열들
columns_to_convert = ['가구소득_(월)금액', '걷기 실천 일수']
for col in columns_to_convert:
    data[col] = pd.to_numeric(data[col], errors='coerce')

# '성별' 기준으로 데이터를 그룹화
grouped_data = data.groupby('성별')

# 변수별로 전체 빈도수와 남자(성별 1), 여자(성별 0)의 빈도수 계산 및 출력
for col in data.columns:
    if col == '성별':  # '성별' 열은 건너뜀
        continue

    print(f'\n{col}의 값별 빈도수 (전체 및 성별):')

    # 전체 빈도수 계산
    total_counts = data[col].value_counts().sort_index()

    # 각 성별 그룹에서 변수별 값의 빈도수 계산
    male_counts = grouped_data[col].value_counts().unstack(fill_value=0).loc[1]
    female_counts = grouped_data[col].value_counts().unstack(fill_value=0).loc[0]

    # 결과 출력
    for value in sorted(data[col].dropna().unique()):
        total_count = total_counts.get(value, 0)
        male_count = male_counts.get(value, 0)
        female_count = female_counts.get(value, 0)
        print(f'값 {value}: {total_count}명 그 중에 남자 {male_count}명, 여자 {female_count}명')
```

### **컬럼별 검진군/비검진군에 따른 남자와 여자 빈도수**

```python
import pandas as pd

# 데이터 변환: 수치형 데이터로 변환할 열들
columns_to_convert = ['가구소득_(월)금액', '걷기 실천 일수']
for col in columns_to_convert:
    data[col] = pd.to_numeric(data[col], errors='coerce')

# '검진군/비검진군 판단'을 기준으로 데이터 나누기 (0: 검진군, 1: 비검진군)
검진군 = data[data['검진군/비검진군 판단'] == 0]
비검진군 = data[data['검진군/비검진군 판단'] == 1]

# 변수별로 검진군과 비검진군에서의 성별 및 혼인 상태 빈도수 계산 및 출력
for col in data.columns:
    if col in ['성별', '검진군/비검진군 판단']:  # '성별'과 '검진군/비검진군 판단' 열은 건너뜀
        continue

    print(f'\n{col}의 검진군과 비검진군 별 빈도수:')

    # 검진군에서의 빈도수
    검진군_total = 검진군[col].value_counts().sort_index()
    검진군_male = 검진군[검진군['성별'] == 1][col].value_counts().sort_index()  # 남자: 성별 1
    검진군_female = 검진군[검진군['성별'] == 0][col].value_counts().sort_index()  # 여자: 성별 0

    # 비검진군에서의 빈도수
    비검진군_total = 비검진군[col].value_counts().sort_index()
    비검진군_male = 비검진군[비검진군['성별'] == 1][col].value_counts().sort_index()  # 남자: 성별 1
    비검진군_female = 비검진군[비검진군['성별'] == 0][col].value_counts().sort_index()  # 여자: 성별 0

    # 결과 출력
    for value in sorted(data[col].dropna().unique()):
        검진군_total_count = 검진군_total.get(value, 0)
        검진군_male_count = 검진군_male.get(value, 0)
        검진군_female_count = 검진군_female.get(value, 0)

        비검진군_total_count = 비검진군_total.get(value, 0)
        비검진군_male_count = 비검진군_male.get(value, 0)
        비검진군_female_count = 비검진군_female.get(value, 0)

        print(f'값 {value}:')
        print(f'  검진군: 총 {검진군_total_count}명 (남자 {검진군_male_count}명, 여자 {검진군_female_count}명)')
        print(f'  비검진군: 총 {비검진군_total_count}명 (남자 {비검진군_male_count}명, 여자 {비검진군_female_count}명)')
```
