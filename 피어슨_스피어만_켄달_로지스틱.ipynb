### **켄달의 타우(p-value)**

```python
# 켄달의 타우 p-value
import pandas as pd
from scipy.stats import kendalltau
import numpy as np

# 지정된 변수 순서
variables = [
    '성별', '연령대', '교육 수준', '일반담배(궐련) 현재흡연 여부', '평생 음주 여부', '주관적 건강수준',
    '뇌졸중_종합', '심근경색_종합', '주관적 스트레스 수준', '우울감 경험 여부',
    '혈압수치 인지 여부', '혈당수치 인지 여부', '걷기 실천 일수', 'BMI(판정기준)', '동/읍면 구분',
    '가구소득_(월)금액', '직업분류(재분류)', '혼인 상태', '검진군/비검진군 판단'  # 마지막에 종속변수 추가
]

# p값을 저장할 행렬 초기화
num_vars = len(variables)
p_value_matrix = np.zeros((num_vars, num_vars))

# 모든 변수 쌍에 대해 kendalltau 계산하여 p값 추출
for i in range(num_vars):
    for j in range(num_vars):
        if i != j:
            _, p_value = kendalltau(data[variables[i]], data[variables[j]])
            p_value_matrix[i, j] = p_value
        else:
            p_value_matrix[i, j] = 0  # 자기 자신과의 p값은 0

# DataFrame으로 변환
p_value_df = pd.DataFrame(p_value_matrix, index=variables, columns=variables)

# 켄달의 타우 p-value 엑셀 파일로 저장
p_value_df.to_excel('kendall_tau_pvalue_최종.xlsx', sheet_name='P-Values Matrix')
```

### **logistic regression(단변량 분석)**

```python
import pandas as pd
import statsmodels.api as sm

# 데이터 로드
data = pd.read_excel("/content/데이터 변환.xlsx")

# 종속 변수 설정
y = data['검진군/비검진군 판단']

# 단변량 로지스틱 회귀 분석을 위한 독립 변수 목록
independent_vars = [
    '성별', '연령대', '교육 수준', '일반담배(궐련) 현재흡연 여부', '평생 음주 여부',
    '주관적 건강수준', '뇌졸중_종합', '심근경색_종합', '주관적 스트레스 수준',
    '우울감 경험 여부', '혈압수치 인지 여부', '혈당수치 인지 여부', '걷기 실천 일수',
    'BMI(판정기준)', '동/읍면 구분', '가구소득_(월)금액', '직업분류(재분류)', '혼인 상태'
]

# 각 독립 변수에 대해 단변량 로지스틱 회귀 분석 수행
for var in independent_vars:
    # 독립 변수 데이터 추출
    X = data[[var]]

    # 상수항 추가
    X = sm.add_constant(X)

    # 로지스틱 회귀 모델 적합
    logit_model = sm.Logit(y, X)
    result = logit_model.fit()

    # 결과 출력
    print(f"독립 변수: {var}")
    print(result.summary())
    print("\n" + "="*80 + "\n")
```

```python
import pandas as pd
import statsmodels.api as sm

# 데이터 로드
data = pd.read_excel("/content/데이터 변환.xlsx")

# 종속 변수 설정
y = data['검진군/비검진군 판단']

# 단변량 로지스틱 회귀 분석을 위한 독립 변수 목록
independent_vars = [
    '성별', '연령대', '교육 수준', '일반담배(궐련) 현재흡연 여부', '평생 음주 여부',
    '주관적 건강수준', '뇌졸중_종합', '심근경색_종합', '주관적 스트레스 수준',
    '우울감 경험 여부', '혈압수치 인지 여부', '혈당수치 인지 여부', '걷기 실천 일수',
    'BMI(판정기준)', '동/읍면 구분', '가구소득_(월)금액', '직업분류(재분류)', '혼인 상태'
]

# 결과를 저장할 리스트
results_list = []

# 각 독립 변수에 대해 단변량 로지스틱 회귀 분석 수행
for var in independent_vars:
    # 독립 변수 데이터 추출
    X = data[[var]]

    # 상수항 추가
    X = sm.add_constant(X)

    # 로지스틱 회귀 모델 적합
    logit_model = sm.Logit(y, X)
    result = logit_model.fit()

    # 결과 저장
    summary_df = pd.DataFrame({
        'Coef.': result.params,
        'Std.Err.': result.bse,
        'z': result.tvalues,
        'P>|z|': result.pvalues,
        '2.5%': result.conf_int()[0],
        '97.5%': result.conf_int()[1]
    })

    # 독립 변수 이름을 추가
    summary_df['Variable'] = var

    # 필요한 열만 선택
    summary_df = summary_df[['Variable', 'Coef.', 'Std.Err.', 'z', 'P>|z|', '2.5%', '97.5%']]

    # 결과 리스트에 추가
    results_list.append(summary_df)

# 모든 결과를 하나의 데이터프레임으로 결합
all_results_df = pd.concat(results_list)

# 엑셀 파일로 저장
all_results_df.to_excel('/content/Univariate_logistic_regression_results.xlsx', index=False)
```
